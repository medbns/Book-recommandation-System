```python
ratings=pd.read_csv('C:/Users/moham/OneDrive/Bureau/data science/BX-Book-Ratings.csv',sep=';',error_bad_lines=False,encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id',
                       'Book-Rating':'rating'},inplace=True)
```

```python
ratings['user_id'].value_counts()
```

```python
ratings['user_id'].unique().shape
```

```python
X=ratings['user_id'].value_counts()>200
```

```python
y=X[X].index
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
books_with_rating=ratings.merge(books,on='ISBN')
```

```python
books_with_rating.head()
```

```python
num_rating=books_with_rating.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.rename(columns={'Title':'title',
                          'rating':'Num_rating'},inplace=True)
```

```python
final_rating=books_with_rating.merge(num_rating, on='title')
```

```python
final_rating.head(2)
```

```python
final_rating=final_rating[final_rating['Num_rating']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[255,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(distance)):
    print(book_pivot.index[suggestion[i]])
```

```python
books_name=book_pivot.index
```

```python
import pickle 
pickle.dump(model,open('C:/Users/moham/OneDrive/Bureau/data science/book_recommander/fold/model.pkl','wb'))
pickle.dump(books_name,open('C:/Users/moham/OneDrive/Bureau/data science/book_recommander/fold/books_name.pkl','wb'))
pickle.dump(final_rating,open('C:/Users/moham/OneDrive/Bureau/data science/book_recommander/fold/final_rating.pkl','wb'))
pickle.dump(book_pivot,open('C:/Users/moham/OneDrive/Bureau/data science/book_recommander/fold/book_pivot.pkl','wb'))
```

```python
def recommand(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distance,suggestion=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(distance)):
        books=book_pivot.index[suggestion[i]]
        for j in books :
            print(j)
    
```

```python
recommand('Year of Wonders')
```
